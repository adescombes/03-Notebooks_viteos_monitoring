```python


def convertir_iso8601(duree_iso):
    try:
        td = isodate.parse_duration(duree_iso)
        # Convertir timedelta en format HH:MM:SS
        total_seconds = int(td.total_seconds())
        heures = total_seconds // 3600
        minutes = (total_seconds % 3600) // 60
        secondes = total_seconds % 60
        return f"{heures:02}:{minutes:02}:{secondes:02}"
    except Exception as e:
        return None  # ou str(e) pour le debug

# Fonction pour parser ISO 8601 en timedelta
def iso_to_timedelta(duree_iso):
    try:
        return isodate.parse_duration(duree_iso)
    except:
        return pd.NaT  # ou pd.Timedelta(0)

```

```python
# convertir format durées

performance_df['engagementDuration'] = performance_df['engagementDuration'].apply(lambda x : iso_to_timedelta(x))
performance_df['drivingDuration'] = performance_df['drivingDuration'].apply(lambda x : iso_to_timedelta(x))
performance_df['drivingBusinessDuration'] = performance_df['drivingBusinessDuration'].apply(lambda x : iso_to_timedelta(x))
performance_df['drivingPrivateDuration'] = performance_df['drivingPrivateDuration'].apply(lambda x : iso_to_timedelta(x))
performance_df['stoppedDuration'] = performance_df['stoppedDuration'].apply(lambda x : iso_to_timedelta(x))
performance_df['idleDuration'] = performance_df['idleDuration'].apply(lambda x : iso_to_timedelta(x))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Colonnes à convertir
duration_cols = [
    "engagementDuration", # engagement = driving + stopped
    "drivingDuration",
    "drivingBusinessDuration",
    "drivingPrivateDuration",
    "stoppedDuration", # temps d'arrêt
    "idleDuration" # temps stationnaire
]

# 2. Conversion en timedelta
for col in duration_cols:
    df[col] = pd.to_timedelta(df[col])

# 3. Conversion en jours
for col in duration_cols:
    df[col + "_days"] = df[col].dt.total_seconds() / (3600 * 24)

# 4. Data pour le graphe
duration_cols_days = [col + "_days" for col in duration_cols]
df_plot = df[["groupName"] + duration_cols_days]

# 5. Format long
df_long = pd.melt(df_plot, id_vars="groupName",
                  var_name="type_duree", value_name="jours")

df_long["type_duree"] = df_long["type_duree"].str.replace("_days", "")
df_long["type_duree"] = df_long["type_duree"].str.replace("Duration", "")

renommage = {
    "engagement": "Utilisation",
    "driving": "Conduite",
    "drivingBusiness": "Conduite pro",
    "drivingPrivate": "Conduite perso",
    "stopped": "Arrêté",
    "idle": "Stationnaire"
}

df_long["type_duree"] = df_long["type_duree"].map(renommage)

# 6. Graphique
plt.figure(figsize=(18, 8))
sns.set_style("white")  # pour retirer le cadre gris
sns.barplot(data=df_long, x="groupName", y="jours", hue="type_duree")

plt.title("Durées de conduite et arrêt - total par groupe du %s au %s" % (formatted_startDate, formatted_endDate))
plt.grid(which="major")

plt.ylabel("Durée (jours)")
plt.xlabel("")  # Pas de nom sur l'axe X
plt.xticks(rotation=45, ha='right')
sns.despine()  # enlève les bordures supérieures/droites

plt.tight_layout()
plt.savefig('../04-Exports/plots/durees_conduite_par_groupe_aout_2025.png')
plt.show()
```
