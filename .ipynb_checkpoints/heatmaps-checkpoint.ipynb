```python
vehicles_df = pd.read_excel('vehicles_details.xlsx')
groups_df = pd.read_excel('groups_details.xlsx')

vehicles_to_group_dict = dict(zip(vehicles_df['id'], vehicles_df['groupId']))
group_to_name_dict = dict(zip(groups_df['id'], groups_df['name']))

journey_df['groupId'] = journey_df['vehicleId'].map(vehicles_to_group_dict)
journey_df['groupName'] = journey_df['groupId'].map(group_to_name_dict)
```

```python
import openrouteservice
from openrouteservice import convert

client = openrouteservice.Client(key = config.get('API_KEY')) 

def get_route(row):
    try:
        route = client.directions(
            coordinates=[(row['startingLongitude'], row['startingLatitude']), (row['endingLongitude'], row['endingLatitude'])],
            profile='driving-car',
            format='geojson'
        )
        return route['features'][0]['geometry']['coordinates']
    except Exception as e:
        return None
```

```python
# /!\ < 2'000 requêtes / jour !!

journey_df['route'] = journey_df.apply(get_route, axis=1)

```

```python
import folium
from folium.plugins import HeatMap
import pandas as pd

# 1. Préparer les données
# Extraire tous les points pour centrer la carte
all_points = []

for route in journey_df['route'].dropna():
    all_points.extend([[lat, lon] for lon, lat in route])  # [lat, lon] for folium

# Calcul centre de carte
if all_points:
    avg_lat = sum(p[0] for p in all_points) / len(all_points)
    avg_lon = sum(p[1] for p in all_points) / len(all_points)
else:
    avg_lat, avg_lon = 48.8566, 2.3522  # Paris fallback

# 2. Créer la carte avec un fond de carte gris
m = folium.Map(location=[avg_lat, avg_lon], zoom_start=12, tiles=None)
folium.TileLayer('CartoDB positron', name='Fond gris clair').add_to(m)

# 3. Créer une heatmap par groupe
for group_name, group_df in journey_df.groupby('groupName'):
    group_points = []
    for route in group_df['route'].dropna():
        group_points.extend([[lat, lon] for lon, lat in route])  # inverser pour folium

    if not group_points:
        continue

    # Créer une couche (FeatureGroup) pour ce groupe
    fg = folium.FeatureGroup(name=str(group_name))

    # Ajouter la heatmap à cette couche
    HeatMap(group_points, radius=6, blur=10).add_to(fg)

    # Ajouter la couche à la carte
    fg.add_to(m)

# 4. Ajouter le contrôle des couches
folium.LayerControl(collapsed=False).add_to(m)

# 5. Exporter la carte
m.save("heatmap_par_groupes.html")
```
